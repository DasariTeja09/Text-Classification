```python
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
```

```python
data=pd.read_csv('/content/Hc3Finalfeatures.csv')
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
```

```python
X_original = df.drop('label',axis=1)
Y_original =  df.label
X_original['Uppercase'] = X_original['text'].str.count(r'[A-Z]')
```

```python
X_text =X_original['text']
```

```python
tokenizer = Tokenizer(num_words=100)
tokenizer.fit_on_texts(X_text)
sequences = tokenizer.texts_to_sequences(X_text)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_sequences_new = pad_sequences(sequences, maxlen=100)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(padded_sequences_new,Y_original,test_size=0.2,random_state=13,stratify=Y_original)
```

```python
padded_sequences = pad_sequences(sequences, maxlen=10)
```

```python
padded_sequences_new.shape
```

```python
model = Sequential([
    Embedding(input_dim=100, output_dim=16, input_length=10),
    Conv1D(32, 5, activation='relu'),
    GlobalMaxPooling1D(),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')  # Binary classification
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train, epochs=10)
```

```python
predictions = model.predict(X_test)
```

```python
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1-score:', f1)
```
