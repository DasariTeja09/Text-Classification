```python
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('/content/Hc3Finalfeatures.csv')
data.info()
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
X_original = df.drop('label',axis=1)
```

```python
Y_original =  df.label
```

adding a feature called uppercase below

```python
X_original['Uppercase'] = X_original['text'].str.count(r'[A-Z]')
```

```python
X_original.head()
```

```python
old_feature_array = X_original.iloc[:, 1:]
old_feature_array.head()
```

```python
X_text =X_original['text']
```

text preprocessing

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
stop_words=set(stopwords.words('english'))
corpus=[]
def cleantxt(text):
  text=re.sub(r'@[A-Za-z0-9]+','',text)
  text=re.sub(r'#','',text)
  text=re.sub(r'http?:\/\/\S+','',text)
  text=text.lower()
  text=text.split()
  text=[x for x in text if not x in stop_words]
  text=[ps.stem(x) for x in text]
  text=" ".join(text)
  corpus.append(text)
  return text

X_text=X_text.apply(cleantxt)
X_text.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectoriser=TfidfVectorizer(max_features=250,min_df=7,max_df=0.8,stop_words=stopwords.words('english'))
pf=vectoriser.fit_transform(corpus).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(new_features_array,Y_original,test_size=0.2,random_state=13,stratify=Y_original)
```

```python
from sklearn.ensemble import RandomForestClassifier
analysis=RandomForestClassifier(n_estimators=50)
analysis.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
pred=analysis.predict(X_test)
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
print(accuracy_score(y_test,pred))
```
